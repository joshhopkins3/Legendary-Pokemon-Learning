```python
#import dependencies
import pandas as pd
import sqlite3

#show all columns in dfs
pd.set_option("display.max_columns", None)
```

```python
#connect to SQLite db
conn = sqlite3.connect('pokedex.db')
```

```python
#read in the csv
pokedex = pd.read_csv('Pokedex_Ver8.csv')
pokedex.head()
```

```python
#clean the dataframe
pokedex.drop(pokedex.columns[23:35], axis=1, inplace=True)
pokedex.drop(columns='MEGA_EVOLUTION_FLAG', inplace=True)
pokedex[['TYPE2', 'ABILITY2', 'ABILITY HIDDEN']] = pokedex[['TYPE2', 'ABILITY2', 'ABILITY HIDDEN']].fillna('None')
pokedex.rename(columns={'ABILITY HIDDEN': 'ABILITY_HIDDEN'}, inplace=True)

pokedex.head()
```

```python
#split dataframe into 8 different dataframes by the generation a pokemon first appeared
pokedex_gens = pokedex.groupby('GENERATION')
gen_1 = pokedex_gens.get_group(1)
gen_2 = pokedex_gens.get_group(2)
gen_3 = pokedex_gens.get_group(3)
gen_4 = pokedex_gens.get_group(4)
gen_5 = pokedex_gens.get_group(5)
gen_6 = pokedex_gens.get_group(6)
gen_7 = pokedex_gens.get_group(7)
gen_8 = pokedex_gens.get_group(8)
```

```python
#combine gens 1 - 6 to create one larger dataframe
gens_1_to_6 = gen_1.append(gen_2).append(gen_3).append(gen_4).append(gen_5).append(gen_6)
gens_1_to_6.head()
```

```python
#save the dataframes as SQL tables in the database
gens_1_to_6.to_sql('gens_1_to_6', conn, index=False)
gen_7.to_sql('gen_7', conn, index=False)
gen_8.to_sql('gen_8', conn, index=False)
```

```python
#close the db connection
conn.close()
```
